Handle Missing Values

```python
# Check for missing values
categorical.isnull().sum()

# Fill missing values
#1. Fill the df['AirBags'] with modal value
# df['AirBags'].fillna(df['AirBags'].mode()[0],inplace=True)

# Fill undefined with Driver-only
df['AirBags'].replace('undefined','Driver-only')

# df['AirBags'].unique()
# Drop missing rows


# Remove duplicates
```

```python
from pandas.api.types import is_numeric_dtype

for i in df.columns:
    # Check if the column is any kind of number
    if is_numeric_dtype(df[i]):
        mean = df[i].mean()
        df[i] = df[i].fillna(mean)
        
    # If it's not a number (so it's a string, object, category, date, etc.)
    else:
        # Check if the column is completely empty to avoid another error
        if not df[i].mode().empty: 
            mode = df[i].mode()[0]
            df[i] = df[i].fillna(mode)
```
